```python
import numpy as np

np.set_printoptions(precision=3, suppress=True)
np.random.seed(42)
```

```python
tokens = ["I", "love", "AI"]

seq_len = len(tokens)
d_model = 2  # small for clarity

# Token embeddings
E = np.array([
    [1.0, 0.0],   # I
    [0.8, 0.6],   # love
    [0.0, 1.0]    # AI
])

print("Embeddings (E):")
print(E)
```

```python
W_Q = np.random.randn(d_model, d_model)
W_K = np.random.randn(d_model, d_model)
W_V = np.random.randn(d_model, d_model)

print("\nW_Q:\n", W_Q)
print("\nW_K:\n", W_K)
print("\nW_V:\n", W_V)
```

```python
Q = E @ W_Q
K = E @ W_K
V = E @ W_V

print("\nQueries (Q):")
print(Q)

print("\nKeys (K):")
print(K)

print("\nValues (V):")
print(V)
```

```python
d_k = Q.shape[-1]

scores = Q @ K.T
scaled_scores = scores / np.sqrt(d_k)

print("\nScaled Attention Scores:")
print(scaled_scores)
```

```python
def softmax(x):
    exp_x = np.exp(x - np.max(x, axis=1, keepdims=True))
    return exp_x / np.sum(exp_x, axis=1, keepdims=True)

attention_weights = softmax(scaled_scores)

print("\nAttention Weights:")
print(attention_weights)

print("\nRow sums (should be 1):")
print(attention_weights.sum(axis=1))
```

```python
context_vectors = attention_weights @ V

print("\nContext Vectors:")
print(context_vectors)
```
