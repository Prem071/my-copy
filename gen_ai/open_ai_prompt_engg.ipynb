```python
from openai import OpenAI

client = OpenAI(api_key="xxxxxxxxx")

# Zero-Shot Prompting

    # Simple tasks

    # Classification

    # Summarization

    # High-level reasoning

response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "user", "content": 
         "Summarize the following customer feedback in one sentence:\n\n"
         "The delivery was late and the packaging was damaged, but customer support eventually resolved the issue."}
    ]
)

print(response.choices[0].message.content)
```

```python
# Few-Shot Prompting

prompt = """
Feedback: "App crashes frequently"
Category: Technical Issue

Feedback: "Delivery was late"
Category: Logistics Issue

Feedback: "Money deducted but payment failed"
Category: Payment Issue

Feedback: "Customer support did not respond"
Category:
"""

response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[{"role": "user", "content": prompt}]
)

print(response.choices[0].message.content)
```

```python
# Chain-of-Thought Prompting (CoT)

response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "user", "content": 
         "A shop sells a product for ₹1200 after giving a 20% discount. "
         "What was the original price? Think step by step."}
    ]
)

print(response.choices[0].message.content)
```

```python
# Role-Based Prompting

response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "system", "content": "You are a Senior Machine Learning Engineer."},
        {"role": "user", "content": 
         "Explain Prompt Engineering to a junior developer using simple examples."}
    ]
)

print(response.choices[0].message.content)
```
