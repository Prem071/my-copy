```python
import numpy as np

np.set_printoptions(precision=3, suppress=True)
np.random.seed(42)
```

```python
tokens = ["I", "love", "AI"]
seq_len = len(tokens)
d_model = 2

# Token embeddings
E = np.array([
    [1.0, 0.0],   # I
    [0.8, 0.6],   # love
    [0.0, 1.0]    # AI
])

print("Embeddings (E):")
print(E)
```

```python
# Learnable weight matrices

W_Q = np.random.randn(d_model, d_model)
W_K = np.random.randn(d_model, d_model)
W_V = np.random.randn(d_model, d_model)

print("\nW_Q:\n", W_Q)
print("\nW_K:\n", W_K)
print("\nW_V:\n", W_V)
```

```python
Q = E @ W_Q
K = E @ W_K
V = E @ W_V

print("\nQueries (Q):")
print(Q)

print("\nKeys (K):")
print(K)

print("\nValues (V):")
print(V)
```

```python
S = Q @ K.T
print("\nSimilarity Scores (QKᵀ):")
print(S)
```

```python
def softmax(x):
    exp_x = np.exp(x - np.max(x, axis=1, keepdims=True))
    return exp_x / np.sum(exp_x, axis=1, keepdims=True)

scaled_S = S / np.sqrt(d_model)

A = softmax(scaled_S)

print("\nAttention Weights:")
print(A)
```

```python
C = A @ V

print("\nContext Vectors:")
print(C)
```
