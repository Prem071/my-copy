```python
from transformers import pipeline

# Zero-Shot Prompting (HF)

generator = pipeline(
    "text-generation",
    model="mistralai/Mistral-7B-Instruct-v0.2",
    max_new_tokens=50
)

prompt = """
Summarize the following feedback in one sentence:
"The delivery was late but customer support resolved it."
"""

output = generator(prompt)
print(output[0]["generated_text"])
```

```python
# Few-Shot Prompting (HF)

prompt = """
Feedback: App crashes frequently
Category: Technical Issue

Feedback: Payment failed but money deducted
Category: Payment Issue

Feedback: Delivery was late
Category:
"""

output = generator(prompt, max_new_tokens=20)
print(output[0]["generated_text"])
```

```python
# Role-Based Prompting (HF)

prompt = """
You are a Senior Data Scientist.
Explain Prompt Engineering to a junior engineer with examples.
"""

output = generator(prompt, max_new_tokens=150)
print(output[0]["generated_text"])
```

```python
# Chain-of-Thought (HF)

prompt = """
A product is sold for ₹1200 after a 20% discount.
What was the original price?
Think step by step.
"""

output = generator(prompt, max_new_tokens=100)
print(output[0]["generated_text"])
```

```python
# Structured Output Prompting (JSON)

prompt = """
Extract sentiment from the feedback.
Return ONLY valid JSON.

Keys:
- sentiment (positive/negative/neutral)
- confidence (0 to 1)

Feedback:
"The product quality is excellent but delivery was slow."
"""

output = generator(prompt, max_new_tokens=80)
print(output[0]["generated_text"])
```

```python
# ReAct Prompting (Agents – HF Style)

prompt = """
You are an intelligent agent.

Task: Determine sentiment.

Thought: Analyze wording
Action: Classify sentiment
Observation: Output result

Feedback: "Refund process was very smooth and fast."
"""

output = generator(prompt, max_new_tokens=120)
print(output[0]["generated_text"])
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch



model_name = "mistralai/Mistral-7B-Instruct-v0.2"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype=torch.float16,
    device_map="auto"
)

prompt = "Explain Prompt Engineering with an example."

inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

outputs = model.generate(
    **inputs,
    max_new_tokens=100,
    temperature=0.7
)

print(tokenizer.decode(outputs[0], skip_special_tokens=True))
```
