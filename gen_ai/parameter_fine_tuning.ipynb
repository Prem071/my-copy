```python
import torch.nn.utils.prune as prune

prune.l1_unstructured(
    module=model.model.layers[0].mlp.fc1,
    name="weight",
    amount=0.3  # prune 30%
)

model.prune_heads({0: [0, 1, 2]})  # prune heads 0,1,2 in layer 0
```

```python
import torch

weight = model.model.layers[0].mlp.fc1.weight
importance = weight.abs().mean(dim=1)

threshold = torch.quantile(importance, 0.3)
mask = importance > threshold

model.model.layers[0].mlp.fc1.weight = torch.nn.Parameter(weight[mask])

model.model.layers = model.model.layers[:-2]  # remove last 2 layers
```
