```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.random.randint(low=25, high=45, size=1000)
```

```python
mean = np.mean(data)
median = np.median(data)

variance = np.var(data)
std_dev  = np.std(data)

print("Mean: ", mean)

```

```python

plt.hist(x=data, bins=2, color='y')
plt.title('Frequency Distribution')
plt.show()
```

```python
symm_data = np.random.normal(scale=1, loc=40, size=1000)

symm_2  = np.random.normal(scale=1, loc=30, size=1000)
symm_3  = np.random.normal(scale=1, loc=36, size=1000)
```

```python
symm_data
```

```python
plt.hist(x=symm_data, bins=15, color='y', edgecolor='black')
plt.grid(True)
plt.title('Normal Distribution')
plt.show()
```

```python
right_skew = np.random.exponential(scale=1, size=1000)
```

```python
plt.hist(x=right_skew, bins=15, color='y')
plt.title('Right Skewed Distribution')
plt.show()
```

```python
left_skew = - np.random.exponential(scale=1, size=1000)
plt.hist(x=left_skew, bins=15, color='y', edgecolor='black')
plt.grid(True)
plt.title('Left Skewed Distribution')
plt.show()
```

```python
plt.boxplot(x=[data])
plt.title('Box Plot')
plt.show()
```

```python
plt.boxplot(x=[symm_data, symm_2, symm_3])
plt.title('Box Plot')
plt.show()
```

```python
plt.violinplot(dataset=[symm_data, symm_2, symm_3])
plt.title('Voilin Plot')
plt.show()
```
