```python
import numpy as np
import pandas as pd

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```

```python
# Load dataset
X, y = load_iris(return_X_y=True)

# Train–test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

```python
from sklearn.naive_bayes import GaussianNB

# Gaussian Naive Bayes assumes:
# x_j | y ~ Normal(mean_{jy}, var_{jy})
nb = GaussianNB()

# MLE happens here
nb.fit(X_train, y_train)
```

```python
# Class predictions (argmax_y P(y|x))
y_pred = nb.predict(X_test)

# Full posterior probabilities P(y|x)
y_proba = nb.predict_proba(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report

print("Accuracy:", round(accuracy_score(y_test, y_pred),2))
print(classification_report(y_test, y_pred))
```

```python
# Inspect learned parameters
print("Class priors P(y):\n", nb.class_prior_)
print("\nMeans (mu_{jy}):\n", nb.theta_)
print("\nVariances (sigma^2_{jy}):\n", nb.var_)
```
