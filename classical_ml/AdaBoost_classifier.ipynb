```python
import pandas as pd
import numpy as np
from sklearn import datasets
```

```python
iris = datasets.load_iris()
input_= pd.DataFrame(data=iris['data'], columns=iris.feature_names)
target_ = pd.DataFrame(data=iris['target'], columns=['type'])
data = pd.concat([input_, target_], axis=1)
data.head(5)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import AdaBoostClassifier, DecisionTreeClassifier

X = data
y = data.pop('type')
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=0.3)

adb = AdaBoostClassifier(
    base_estimator=DecisionTreeClassifier(
        max_depth=2,
        min_samples_leaf=5
    ),
    n_estimators=100,
    learning_rate=0.1,
    algorithm="SAMME.R",
    random_state=42
)

model = adb.fit(X_train, Y_train)
y_pred = model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
print(confusion_matrix(Y_test, y_pred))
print(classification_report(Y_test, y_pred))
print(accuracy_score(Y_test, y_pred))
```
