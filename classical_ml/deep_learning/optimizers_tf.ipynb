```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
tf.random.set_seed(42)
np.random.seed(42)
```

```python
X = np.random.randn(1000, 1)
y = 3 * X + 0.5 * np.random.randn(1000, 1)

X_train, y_train = X[:800], y[:800]
X_val, y_val = X[800:], y[800:]
```

```python
def build_model():
    model = tf.keras.Sequential([
        tf.keras.layers.Dense(1, input_shape=(1,))
    ])
    return model


def train_with_optimizer(optimizer, epochs=50):
    model = build_model()
    model.compile(
        optimizer=optimizer,
        loss="mse"
    )
    
    history = model.fit(
        X_train, y_train,
        validation_data=(X_val, y_val),
        epochs=epochs,
        verbose=0
    )
    return history
```

```python
opt_sgd = tf.keras.optimizers.SGD(learning_rate=0.01)

opt_momentum = tf.keras.optimizers.SGD(
    learning_rate=0.01,
    momentum=0.9
)

opt_adagrad = tf.keras.optimizers.Adagrad(learning_rate=0.1)

opt_rmsprop = tf.keras.optimizers.RMSprop(
    learning_rate=0.01,
    rho=0.9
)

opt_adam = tf.keras.optimizers.Adam(learning_rate=0.01)

```

```python
histories = {}
plt.figure(figsize=(10,6))

for name, history in histories.items():
    plt.plot(history.history["loss"], label=name)

plt.yscale("log")
plt.xlabel("Epoch")
plt.ylabel("Training Loss (log scale)")
plt.title("Optimizer Comparison – Training Loss")
plt.legend()
plt.show()

histories["SGD"] = train_with_optimizer(opt_sgd)
histories["Momentum"] = train_with_optimizer(opt_momentum)
histories["AdaGrad"] = train_with_optimizer(opt_adagrad)
histories["RMSProp"] = train_with_optimizer(opt_rmsprop)
histories["Adam"] = train_with_optimizer(opt_adam)
```

```python
plt.figure(figsize=(10,6))

for name, history in histories.items():
    plt.plot(history.history["loss"], label=name)

plt.yscale("log")
plt.xlabel("Epoch")
plt.ylabel("Training Loss (log scale)")
plt.title("Optimizer Comparison – Training Loss")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,6))

for name, history in histories.items():
    plt.plot(history.history["val_loss"], label=name)

plt.yscale("log")
plt.xlabel("Epoch")
plt.ylabel("Validation Loss (log scale)")
plt.title("Optimizer Comparison – Validation Loss")
plt.legend()
plt.show()
```
