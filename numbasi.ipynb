```python
#Finding the angle between two vectors

a = np.array([1,0])
b = np.array([0,1])
```

```python
plt.figure(figsize=(8,6))
plt.quiver(0,0,*a,scale_units='xy', angles="xy",scale=1, color='r', label='v')
plt.quiver(0,0,*b,scale_units='xy', angles="xy",scale=1, color='b', label='v')
plt.title('Visualizing an orthogonal Vector')
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.xlim(-1,1.5)
plt.ylim(-1,1.5)
plt.grid(True)
plt.show()
```

```python
np.degrees(np.arccos(np.dot(a,b)/(np.linalg.norm(a) * np.linalg.norm(b))))
```

```python
a = np.array([2,0])
b = np.array([1,1])

proj_b = (np.dot(a,b)/np.dot(a,a)) * a

plt.figure(figsize=(8,6))
plt.quiver(0,0,*a,scale_units='xy', angles="xy",scale=1, color='r', label='v')
plt.quiver(0,0,*b,scale_units='xy', angles="xy",scale=1, color='b', label='v')
plt.quiver(0,0,*proj_b,scale_units='xy', angles="xy",scale=1, color='g', label='v')
plt.title('Visualizing Vector Projection')
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.xlim(-1,3.5)
plt.ylim(-1,3.5)
plt.grid(True)
plt.show()
```
